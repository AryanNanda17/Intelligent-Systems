# Importing Matplotlib 

To use Matplotlib in your Python program, the standard way is:

````python
import matplotlib.pyplot as plt
````

Here, 
- import matplotlib: This loads the Matplotlib library, which is used for data visualization.
- pyplot: This submodule of Matplotlib is used for creating 2D plots.
- as plt: This creates an alias (plt) for Matplotlib’s pyplot module, making the code concise.


#### Direct Import
You can also import all functions from Matplotlib directly:

````python
from matplotlib.pyplot import *
````

Here’s an example of using Matplotlib to create a simple line plot:

````python
# Importing Matplotlib
import matplotlib.pyplot as plt

# Creating data for the plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# Creating a line plot
plt.plot(x, y)

# Adding a title and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Displaying the plot
plt.show()
````

## Basic Graph

````python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graph
x = [0, 1, 2, 3, 4, 5]
y = [0, 3, 6, 9, 12, 15]

# Resize the graph (dpi controls resolution (number of pixels per inch))
plt.figure(figsize=(5, 3), dpi=120)

# Line 1
# Using keyword arguments for customization
plt.plot(x, y, label='3x', color='green', linewidth=2.5, marker='o', linestyle='-', markersize=8, markeredgecolor='black')

# Line 2
# Generate finer intervals
x2 = np.linspace(0, 5, 20) # 20 equally separated numbers between 0 and 5 (inclusive) 
# print(x2, x2**1.5)
# Plot using shorthand notation for style
plt.plot(x2, x2**1.5, 'm--', label='X^1.5')

# Line 3
# Different data for demonstration
x3 = np.linspace(0, 5, 15)
plt.plot(x3, np.sqrt(x3)*5, 'c-.', label='5√X')

# Add a title with different font properties
plt.title('A Customized Graph', fontdict={'fontname': 'Georgia', 'fontsize': 22, 'color': 'purple'})

# X and Y labels
plt.xlabel('Input Values', fontsize=12)
plt.ylabel('Output Values', fontsize=12)

# Custom tick marks for X-axis and Y-axis
plt.xticks(range(0, 6))
plt.yticks([0, 5, 10, 15, 20, 25])

# Add a grid for better readability
plt.grid(alpha=0.3)

# Add a legend to differentiate lines
plt.legend()

# Save the figure (high resolution for clarity)
# plt.savefig('customized_graph.png', dpi=300)

# Display the plot
plt.show()
````

## Shorthand notation

Shorthand notation in Matplotlib allows you to specify line style, marker style, and line color in a single string when plotting, instead of using separate keyword arguments for each. This makes the code concise and easy to write. 

Format of Shorthand Notation:
The shorthand string follows the format:

'[color][marker][line]'

Each component is optional, so you can use just one or a combination of two or three. Here's a breakdown:

1. Color ([color])
   
Specifies the color of the line and markers. Common options include:

- 'b' = Blue
- 'g' = Green
- 'r' = Red
- 'c' = Cyan
- 'm' = Magenta
- 'y' = Yellow
- 'k' = Black
- 'w' = White

Example:
```python
'k'  # Black color
'r'  # Red color
```

2. Marker ([marker])
   
Specifies the style of the marker to indicate individual data points. Common options include:

- '.' = Point
- 'o' = Circle
- 'x' = X marker
- '+' = Plus
- '^' = Upward triangle
- 'v' = Downward triangle
- 's' = Square
- '*' = Star
- 'd' = Diamond

Example:

```python
'o'  # Circle markers
'x'  # X markers
```

3. Line Style ([line])
   
Specifies the style of the line connecting the data points. Common options include:


- '-' = Solid line
- '--' = Dashed line
- '-.' = Dash-dot line
- ':' = Dotted line


Example:

```python
'--'  # Dashed line
':'   # Dotted line
```
