# Sorting

Key Options:
- by: Column(s) to sort by.
- ascending: Set to True for ascending order (default) or False for descending orde

```python
# Sort by a single column (e.g., "HP") in ascending order
sorted_df = df.sort_values(by="HP")
sorted_df
```

```python
# Sort by "Attack" in descending order
sorted_df = df.sort_values(by="Attack", ascending=False)
sorted_df
```

```python
# Sort by multiple columns
sorted_df = df.sort_values(by=["Type 1", "HP"], ascending=[True, False])
sorted_df
```

```python
grass_pokemon = data[data['Type 1'] == 'Grass']
grass_pokemon.head()
```

## Filter by Multiple Conditions

```python
fire_strong_pokemon = data[(data['Type 1'] == 'Fire') & (data['Attack'] > 70)]
fire_strong_pokemon.head()
```

## Filter Using isin

```python
fire_grass_pokemon = data[data['Type 1'].isin(['Fire', 'Grass'])]
fire_grass_pokemon.head()
```

```python
missing_type2 = data[data['Type 2'].isnull()]
missing_type2.head()
```

## Filter by Range of Values

```python
medium_hp_pokemon = data[(data['HP'] >= 50) & (data['HP'] <= 100)]
medium_hp_pokemon.head()
```

# Conditional Changes

Conditional changes in Pandas involve updating the values of certain rows in the DataFrame based on specific conditions. This is typically done using boolean indexing or the loc method.

```python
import pandas as pd

# Reading the data into a DataFrame
data = pd.read_csv("pokemon_data.csv")
```

## Change a Column's Value Conditionally

Let's say you want to mark all Pokémon with Type 1 as "Fire" as "Hot" in a new column Category.

```python
# Add a new column with default value
data['Category'] = 'Normal'

# Update 'Category' to 'Hot' where 'Type 1' is 'Fire'
data.loc[data['Type 1'] == 'Fire', 'Category'] = 'Hot'

data[['Name', 'Type 1', 'Category']].head()
```

## Modify Multiple Columns Conditionally
If you want to boost the Attack of all Fire Pokémon by 10 and set their Legendary status to True

```python
# Update 'Attack' and 'Legendary' conditionally
data.loc[data['Type 1'] == 'Fire', 'Attack'] += 10
data.loc[data['Type 1'] == 'Fire', 'Legendary'] = True

# Display the relevant columns
print(data[['Name', 'Type 1', 'Attack', 'Legendary']].head())
```

## Use Conditional Changes with Multiple Conditions
If you want to mark Pokémon as "Tank" in the Category column when their Defense is greater than 100 and their HP is greater than 80:

```python
# Update 'Category' to 'Tank' based on multiple conditions
data.loc[(data['Defense'] > 100) & (data['HP'] > 80), 'Category'] = 'Tank'

data[['Name', 'Defense', 'HP', 'Category']].head()
```
