```python
# Import pandas
import pandas as pd
```

```python
df = pd.read_csv('pokemon_data.csv')
```

```python
df
```

# Viewing Data

```python
# View the First N Rows:
# head() returns the first N rows (default is 5).
df.head()  # First 5 rows
```

```python
df.head(10)  # First 10 rows
```

```python
# View the Last N Rows:
# tail() returns the last N rows (default is 5).
df.tail()  # Last 5 rows
```

```python
df.tail(10)  # Last 10 rows
```

```python
# View the Shape of the DataFrame:
# Returns the number of rows and columns in the DataFrame.
df.shape  # (number of rows, number of columns)
```

```python
# View DataFrame Information:
# info() provides a summary of the DataFrame, including column types and non-null counts.
df.info()
```

```python
# View Summary Statistics:
# describe() gives statistics (mean, std, min, max) for numeric columns.
df.describe()
```

```python
# View Column Names:
# columns shows all the column names.
df.columns
```

```python
# View Row Labels/Index:
# index shows the row labels (index).
df.index
```

```python
# View Data Types of Columns:
# dtypes shows the data type of each column.
df.dtypes
```

## Advanced Viewing

```python
# View a Random Sample:
# sample() returns a random sample of rows from the DataFrame.
df.sample(3)  # 3 random rows
```

```python
# View Unique Values in a Column:
# unique() returns the unique values of a specific column.
df['Type 1'].unique()
```

```python
# View Frequency of Unique Values in a Column:
# value_counts() returns the count of unique values in a column
df['Type 1'].value_counts()
```
