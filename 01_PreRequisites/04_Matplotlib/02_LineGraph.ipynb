# Line Graph

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Stock Price Analysis

```python
stocks = pd.read_csv('stocks_prices.csv')

stocks
```

```python
plt.figure(figsize=(8,5))

plt.title('stocks Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(stocks.Year, stocks.Apple, 'b.-', label='Apple')
plt.plot(stocks.Year, stocks.Microsoft, 'r.-', label='Microsoft')
plt.plot(stocks.Year, stocks.Tesla, 'g.-', label='Tesla')
plt.plot(stocks.Year, stocks.Amazon, 'y.-', label='Amazon')

# Another Way to plot many values!
# countries_to_look_at = ['Apple', 'Microsoft', 'Tesla', 'Amazon']
# for stock in stocks:
#     if country in countries_to_look_at:
#         plt.plot(stocks.Year, stocks[stock], marker='.')

plt.xticks(stocks.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('stocks_price_figure.png', dpi=300)

plt.show()
```

## Gas Price Analysis

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
plt.plot(gas.Year, gas.Canada, 'r.-')
plt.plot(gas.Year, gas['South Korea'], 'g.-')
plt.plot(gas.Year, gas.Australia, 'y.-')

# Another Way to plot many values!
# countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```
